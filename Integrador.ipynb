Es una medida que indica cuanta energia primaria se necesita para producir una unidad de Producto Bruto Interno (PBI) de una economía. Refleja la eficiencia con la que una economía utiliza energia para generar riqueza, Una relación mas baja significa que la economía utiliza menos energía para producir una unidad de PIB, lo que indica mayor eficiencia energetica.

10. Producción de energia renovable:
